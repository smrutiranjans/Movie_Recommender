```python
#popularity based
#content based
#collaborative
#hybrid
import pandas as pd
import numpy as np
movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
```

```python
new_df=movies.merge(credits,on='title')
```

```python
new_df=new_df[[ 'title','genres', 'keywords', 'overview', 'movie_id', 'cast', 'crew']]
```

```python
new_df.shape
```

```python
new_df
```

```python
import ast
new_df.loc[0,'genres']
```

```python
def convert(genres):
    l=[]
    for i in ast.literal_eval(genres):
        l.append(i['name'])
    return l
```

```python

new_df['genres']=new_df['genres'].apply(convert)
```

```python
new_df.isnull().sum()
```

```python
new_df.dropna(inplace=True)
```

```python
new_df
```

```python
new_df.loc[0,'keywords']
```

```python
new_df['keywords']=new_df['keywords'].apply(convert)
```

```python
new_df.loc[0,'overview']
```

```python
new_df.loc[0,'cast']
```

```python
def cast(casts):
    l=[]
    c=0
    for i in ast.literal_eval(casts):
        if c!=3:
            l.append(i['name'].replace(' ',''))
            c+=1
        else:
            break
    return l
        
```

```python
new_df['cast']=new_df['cast'].apply(cast)
```

```python
new_df.loc[0,'crew']
```

```python
def director(crews):
    l=[]
    for i in ast.literal_eval(crews):
        if i['job']=='Director':
            l.append(i['name'].replace(' ',''))
    return l
```

```python
new_df['crew']=new_df['crew'].apply(director)
new_df
```

```python
new_df['tags']=new_df['genres']+new_df['keywords']+new_df['overview'].apply(lambda x:x.split())+new_df['cast']+new_df['crew']
```

```python
new_df.iloc[0,-1]
```

```python
new_df.drop(columns=['genres','overview','keywords','cast','crew'],inplace=True)
```

```python
new_df.drop(columns=['movie_id'],inplace=True)
new_df
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:' '.join(x))
new_df
```

```python
new_df.iloc[0,1]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
vector=cv.fit_transform(new_df['tags']).toarray()
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
new_df[new_df['title']=='The Dark Knight Rises'].index[0]
```

## Logic of recommendation

```python
'''
1.user inputs a movie name
2.we take the values and check for the top 5 most related movies
3.assoiate index values to the movies
'''
def recommend(movie):
    index=new_df[new_df['title']==movie].index[0] 
    distances=sorted(list(enumerate(similarity[index])),reverse=True,key=lambda x:x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
```

```python
dist=sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])
dist
```

```python
recommend('The Dark Knight Rises')
```
